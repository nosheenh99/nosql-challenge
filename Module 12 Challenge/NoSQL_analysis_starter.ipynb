### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

- Search within 0.01 degree on either side of the latitude and longitude.
- Rating value must equal 5
- Sort by hygiene score

```python
# Find the Penang Flavours document
penang_flavours = db.establishments.find_one({"BusinessName": "Penang Flavours"})

# Get the existing latitude and longitude values
latitude = penang_flavours["geocode"]["latitude"]
longitude = penang_flavours["geocode"]["longitude"]

# Update the document with Decimal128 data type for latitude and longitude
db.establishments.update_one(
    {"BusinessName": "Penang Flavours"},
    {
        "$set": {
            "geocode.latitude": Decimal128(str(latitude)),
            "geocode.longitude": Decimal128(str(longitude))
        }
    }
)
```

```python
degree_search = 0.01

penang_flavours_data = db.establishments.find_one({"BusinessName": "Penang Flavours"})

latitude = penang_flavours_data['geocode']['latitude']
longitude = penang_flavours_data['geocode']['longitude']
```

```python
query = {
    "geocode.latitude": {"$gte": Decimal128(str(float(str(latitude)) - degree_search)), "$lte": Decimal128(str(float(str(latitude)) + degree_search))},
    "geocode.longitude": {"$gte": Decimal128(str(float(str(longitude)) - degree_search)), "$lte": Decimal128(str(float(str(longitude)) + degree_search))},
    "RatingValue": 5,
}
```

```python
sort = [("scores.Hygiene", 1)]
```

```python
# Print the results
result = db.establishments.find(query).sort(sort).limit(5)
```

```python
# Print the results
for res in result:
    pprint(res)
```

```python
# Convert result to Pandas DataFrame
df = pd.DataFrame(list(db.establishments.find(query).sort(sort).limit(5)))
df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

#### Create a pipeline that: 
- 1. Matches establishments with a hygiene score of 0
- 2. Groups the matches by Local Authority
- 3. Sorts the matches from highest to lowest

```python
pipeline = [
    # Match establishments with a hygiene score of 0
    {"$match": {"scores.Hygiene": 0}},
    
    # Group the matches by Local Authority and count them
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    
    # Sort the matches from highest to lowest
    {"$sort": {"count": -1}},
]

# Execute the pipeline
result = db.establishments.aggregate(pipeline)
result = list(result)
```

```python
# Print the number of documents in the result
print(len(result))
```

```python
# Print the first 10 results
for document in result[:10]:
    pprint(document)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(result)
```

```python
# Display the number of rows in the DataFrame
print(f'Number of rows: {df.shape[0]}')
```

```python
# Display the first 10 rows of the DataFrame
df.head(10)
```
